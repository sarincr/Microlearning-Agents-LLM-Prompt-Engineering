```python
import subprocess
import time
import threading
import ollama

# 1. Setup (Your existing server logic)
def run_ollama():
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

threading.Thread(target=run_ollama, daemon=True).start()
time.sleep(5)

# 2. Our "Private Knowledge Base"
# In a real RAG, this would be a PDF or a Database.
documents = [
    "The secret ingredient in the company cafeteria's soup is toasted pumpkin seeds.",
    "The office printer is named 'Bob' and he prefers 80gsm paper.",
    "Employees are allowed to bring their pet iguanas to work on Fridays."
]

def get_relevant_context(query, docs):
    """
    A very simple 'search' function.
    In production, you'd use embeddings and a vector DB here.
    """
    return "\n".join([doc for doc in docs if any(word in doc.lower() for word in query.lower().split())])

# 3. The RAG Logic
query = "What is the secret ingredient in the soup?"
context = get_relevant_context(query, documents)

# 4. Prompt (Combining context + query)
prompt = f"""
Use the following pieces of context to answer the question.
If the answer isn't in the context, say you don't know.

Context: {context}
Question: {query}
"""

print(f"--- Querying with Context ---\n")
response = ollama.chat(model='llama3.2', messages=[
    {'role': 'user', 'content': prompt}
])

print(response['message']['content'])
```

```python
# =========================================================
# INSTALL (if needed)
# =========================================================
# !pip install -U langchain langchain-core langchain-community chromadb sentence-transformers

# =========================================================
# IMPORTS
# =========================================================
import os
import time
from functools import lru_cache

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_community.chat_models import ChatOllama

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

from sentence_transformers import CrossEncoder


# =========================================================
# CONFIG
# =========================================================
TOP_K_INITIAL = 6      # lower = faster
TOP_K_FINAL = 3
NUM_AUG_QUERIES = 1   # reduce LLM calls → faster
RERANK_BATCH = 8


# =========================================================
# HR POLICY TEXT
# =========================================================
hr_policy_text = """
HUMAN RESOURCES POLICY

Maternity leave is 26 weeks.
Paternity leave is 2 weeks.
Employees can report grievances to HR.
Working hours are 9 AM to 6 PM.
"""


# =========================================================
# SPLIT TEXT
# =========================================================
splitter = RecursiveCharacterTextSplitter(
    chunk_size=350,
    chunk_overlap=100
)
docs = splitter.create_documents([hr_policy_text])


# =========================================================
# EMBEDDINGS (FAST LOCAL)
# =========================================================
embedder = SentenceTransformerEmbeddings(
    model_name="all-MiniLM-L6-v2"
)


# =========================================================
# CHROMA VECTOR STORE
# =========================================================
persist_dir = "chroma_db"

if os.path.exists(persist_dir):
    vectorstore = Chroma(
        persist_directory=persist_dir,
        embedding_function=embedder
    )
else:
    vectorstore = Chroma.from_documents(
        documents=docs,
        embedding=embedder,
        persist_directory=persist_dir
    )


# =========================================================
# RETRIEVER
# =========================================================
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={"k": TOP_K_INITIAL, "lambda_mult": 0.7}
)


# =========================================================
# LOCAL LLM (OLLAMA)
# =========================================================
llm = ChatOllama(
    model="llama3.2",   # make sure you pulled this
    base_url="http://127.0.0.1:11434"
)


# =========================================================
# MULTI-QUERY (CACHED)
# =========================================================
multi_query_prompt = ChatPromptTemplate.from_template("""
Generate {n} alternative search queries.
Question: {question}
""")

@lru_cache(maxsize=128)
def generate_queries(question):
    response = llm.invoke(
        multi_query_prompt.format_messages(question=question, n=NUM_AUG_QUERIES)
    ).content

    return [q.strip() for q in response.split("\n") if q.strip()]


# =========================================================
# RERANKER (BATCHED)
# =========================================================
reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

def rerank(question, docs):
    pairs = [(question, d.page_content) for d in docs]

    scores = []
    for i in range(0, len(pairs), RERANK_BATCH):
        batch = pairs[i:i+RERANK_BATCH]
        scores.extend(reranker.predict(batch))

    ranked = sorted(zip(docs, scores), key=lambda x: x[1], reverse=True)
    return [doc for doc, _ in ranked[:TOP_K_FINAL]]


# =========================================================
# RETRIEVE + RERANK
# =========================================================
def retrieve(question):
    queries = [question] + generate_queries(question)

    all_docs = []
    for q in queries:
        all_docs.extend(retriever.invoke(q))

    unique_docs = list({d.page_content: d for d in all_docs}.values())

    return rerank(question, unique_docs)


# =========================================================
# FORMAT CONTEXT
# =========================================================
def format_docs(docs):
    return "\n\n".join(
        [f"[Source {i+1}]\n{doc.page_content}" for i, doc in enumerate(docs)]
    )


# =========================================================
# PROMPT
# =========================================================
prompt = ChatPromptTemplate.from_template("""
You are an HR assistant.

- Answer ONLY from context
- If not found, say "I don't know"
- Include sources

Context:
{context}

Question:
{question}
""")


# =========================================================
# ANSWER PIPELINE
# =========================================================
def get_answer(question):
    start = time.time()

    docs = retrieve(question)
    context = format_docs(docs)

    chain = prompt | llm | StrOutputParser()

    answer = chain.invoke({
        "context": context,
        "question": question
    })

    latency = time.time() - start
    return answer, docs, latency


# =========================================================
# EVALUATION
# =========================================================
def precision_at_k(docs, keywords):
    hits = sum(1 for d in docs if any(k in d.page_content.lower() for k in keywords))
    return hits / len(docs) if docs else 0

def mrr(docs, keywords):
    for i, d in enumerate(docs, 1):
        if any(k in d.page_content.lower() for k in keywords):
            return 1 / i
    return 0


# =========================================================
# CHAT LOOP
# =========================================================
print("🤖 Local HR Chatbot Ready (type 'exit')\n")

while True:
    q = input("You: ").strip()

    if q.lower() == "exit":
        break

    if not q:
        continue

    answer, docs, latency = get_answer(q)

    if len(answer.strip()) < 5:
        answer = "I don't know based on policy."

    print("\nBot:", answer)
    print("⏱ Latency:", round(latency, 2), "sec")

    keywords = q.lower().split()
    print("📊 Precision@k:", round(precision_at_k(docs, keywords), 2))
    print("📊 MRR:", round(mrr(docs, keywords), 2))
    print()
```

```python
# =========================================================
# INSTALL (run once)
# =========================================================
# !pip install -U langchain langchain-ollama langchain-chroma \
# langchain-huggingface sentence-transformers chromadb nest_asyncio

# =========================================================
# FIX ASYNC (CRITICAL FOR NOTEBOOKS)
# =========================================================
import nest_asyncio
nest_asyncio.apply()

# =========================================================
# IMPORTS
# =========================================================
import asyncio, time, sys
from functools import lru_cache

from langchain_ollama import ChatOllama
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import CrossEncoder


# =========================================================
# CONFIG
# =========================================================
MAX_STEPS = 1
TOP_K = 4
RERANK_BATCH = 16


# =========================================================
# LLM (OLLAMA)
# =========================================================
llm = ChatOllama(
    model="llama3.2",
    temperature=0,
    num_predict=128,
)


# =========================================================
# STREAMING OUTPUT
# =========================================================
def stream_print(text):
    for char in text:
        sys.stdout.write(char)
        sys.stdout.flush()
        time.sleep(0.003)


# =========================================================
# DATA
# =========================================================
text = """
Maternity leave is 26 weeks.
Paternity leave is 2 weeks.
Employees can report grievances to HR.
Working hours are 9 AM to 6 PM.
"""

splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=80)
docs = splitter.create_documents([text])

embedder = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vectorstore = Chroma.from_documents(docs, embedder)
retriever = vectorstore.as_retriever(search_kwargs={"k": TOP_K})


# =========================================================
# RERANKER (FAST BATCH)
# =========================================================
reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

def rerank(q, docs):
    pairs = [(q, d.page_content) for d in docs]
    scores = []

    for i in range(0, len(pairs), RERANK_BATCH):
        scores.extend(reranker.predict(pairs[i:i+RERANK_BATCH]))

    ranked = sorted(zip(docs, scores), key=lambda x: x[1], reverse=True)
    return [d for d, _ in ranked[:3]]


# =========================================================
# FAST RETRIEVAL (ASYNC)
# =========================================================
async def fast_retrieve(q):
    docs = retriever.invoke(q)
    docs = rerank(q, docs)
    return "\n\n".join([d.page_content for d in docs])


# =========================================================
# MEMORY
# =========================================================
chat_history = []

def get_memory():
    return "\n".join(chat_history[-4:])


# =========================================================
# TOOL SYSTEM
# =========================================================
async def tool_search(q):
    return await fast_retrieve(q)

TOOLS = {
    "search": tool_search
}


# =========================================================
# ROUTER (ADAPTIVE)
# =========================================================
def route(q):
    res = llm.invoke(f"""
Decide FAST or AGENT.

Question: {q}
""").content.strip()
    return res


# =========================================================
# REACT EXECUTOR
# =========================================================
async def executor(q):
    scratchpad = ""

    for _ in range(MAX_STEPS):
        prompt = f"""
You are an agent.

Tools: search

Question: {q}

{scratchpad}

Respond:
Action:
Action Input:
"""

        res = llm.invoke(prompt).content

        if "search" in res:
            obs = await tool_search(q)
            scratchpad += f"\nObservation: {obs}\n"
        else:
            return res

    return "I don't know"


# =========================================================
# SELF REFLECTION
# =========================================================
def reflect(answer, q):
    return llm.invoke(f"""
Is this answer correct?

Answer: {answer}
Question: {q}

Return OK or FIX
""").content.strip()


# =========================================================
# FAST PATH (RAG)
# =========================================================
async def fast_answer(q):
    context = await fast_retrieve(q)

    return llm.invoke(f"""
Context:
{context}

History:
{get_memory()}

Question:
{q}
""").content


# =========================================================
# SAFE ASYNC RUNNER
# =========================================================
def run_async(coro):
    return asyncio.get_event_loop().run_until_complete(coro)


# =========================================================
# MAIN LOOP
# =========================================================
print("⚡ Ultra-Fast Hybrid Agent Ready\n")

while True:
    q = input("You: ")

    if q.lower() == "exit":
        break

    start = time.time()

    mode = route(q)

    if "FAST" in mode:
        answer = run_async(fast_answer(q))
    else:
        answer = run_async(executor(q))

    # reflection (lightweight)
    if len(answer) < 30:
        if reflect(answer, q) != "OK":
            answer = run_async(fast_answer(q))

    latency = time.time() - start

    print("\nBot: ", end="")
    stream_print(answer)

    print(f"\n⏱ {round(latency,2)} sec\n")

    chat_history.append(f"User: {q}")
    chat_history.append(f"Bot: {answer}")
```
