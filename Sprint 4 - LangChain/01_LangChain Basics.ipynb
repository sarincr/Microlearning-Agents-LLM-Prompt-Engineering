```python
# 1. Install Ollama and the Python library
!curl -fsSL https://ollama.com/install.sh | sh
!pip install ollama -q

import subprocess
import time
import threading
import ollama

# 2. Start Ollama server in the background
def run_ollama():
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

threading.Thread(target=run_ollama, daemon=True).start()
time.sleep(5)  # Give the server a moment to warm up

# 3. Pull a odel (Llama 3.2 is ~2GB)
print("Downloading model...")
ollama.pull('llama3.2')

# 4. Ask a simple question
print("\n--- Model Response ---")
response = ollama.chat(model='llama3.2', messages=[
    {'role': 'user', 'content': 'What is the most interesting fact about space?'}
])

print(response['message']['content'])
```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 1. Setup the Components
llm = OllamaLLM(model="llama3.2")
parser = StrOutputParser()

# 2. Define a Template
# Use {topic} as a placeholder
prompt = ChatPromptTemplate.from_template("Tell me a short, funny joke about {topic}.")

# 3. Create the Chain using the pipe (|) operator
# Flow: Input -> Prompt -> Model -> String Parser
chain = prompt | llm | parser

# 4. Run the Chain
topic = "robots"
response = chain.invoke({"topic": topic})

print(f"--- Joke about {topic} ---")
print(response)
```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

# 1. Setup the Model (Existing server)
llm = OllamaLLM(model="llama3.2")

# 2. Define a Template
prompt = ChatPromptTemplate.from_template("What is the chemical symbol for {element}?")

# 3. Create the Chain WITHOUT a parser
# Data flow stops at the LLM
raw_chain = prompt | llm

# 4. Run it
response = raw_chain.invoke({"element": "Gold"})

print("--- Raw Output (No Parser) ---")
print(f"Type: {type(response)}") # This will be a simple <class 'str'>
print(f"Content: {response}")
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a specialized {role} trainer. Use professional but accessible language."),
    ("user", "Explain how {concept} applies to our workflow.")
])

# Formatting with multiple variables
formatted_chat = chat_template.format_messages(
    role="Generative AI",
    concept="Tokenization"
)

for message in formatted_chat:
    print(f"[{message.type.upper()}]: {message.content}")
```

```python
from langchain_core.prompts import FewShotPromptTemplate, PromptTemplate
from langchain_ollama import OllamaLLM

# 1. Provide the "Examples" (The Pattern)
examples = [
    {"word": "Happy", "antonym": "Sad"},
    {"word": "Tall", "antonym": "Short"}
]

# 2. Tell LangChain how to format each example
example_formatter = PromptTemplate.from_template("Word: {word} -> Antonym: {antonym}")

# 3. Create the "Few-Shot" Blueprint
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_formatter,
    suffix="Word: {input} -> Antonym:", # This is where the new question goes
    input_variables=["input"]
)

# 4. Run it through your Llama model
llm = OllamaLLM(model="llama3.2")
chain = few_shot_prompt | llm

print(chain.invoke({"input": "Hot"}))
# Expected Output: Cold
```
