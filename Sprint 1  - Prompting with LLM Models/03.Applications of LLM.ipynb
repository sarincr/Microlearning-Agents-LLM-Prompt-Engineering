```python
import time
from transformers import pipeline

# --- APPROACH 1: WITHOUT AI (KEYWORD MATCHING) ---
# This is fast but "brittle" (fails if the user uses synonyms).
def detect_intent_keywords(text):
    text = text.lower()
    if any(word in text for word in ["order", "track", "delivery", "package"]):
        return "Order_Status"
    elif any(word in text for word in ["cancel", "refund", "return", "money"]):
        return "Refund_Request"
    elif any(word in text for word in ["login", "password", "account", "access"]):
        return "Account_Security"
    else:
        return "Fallback_Intent"

# --- APPROACH 2: WITH AI (TRANSFORMERS) ---
# This understands the "meaning" behind the words.
# It downloads about 1.6GB of data on first run.
intent_model = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def detect_intent_ai(text):
    labels = ["Order_Status", "Refund_Request", "Account_Security", "Greeting"]
    result = intent_model(text, candidate_labels=labels)
    return result['labels'][0], result['scores'][0]

# --- TEST DATA ---
# This query does NOT contain the word 'password' or 'order'
test_query = "I am having trouble getting into my dashboard."

print("--- RESULTS COMPARISON ---")

# Test Keyword Version
start = time.time()
kw_intent = detect_intent_keywords(test_query)
print(f"1. Keyword Result: {kw_intent} (Time: {time.time()-start:.4f}s)")

# Test AI Version
start = time.time()
ai_intent, score = detect_intent_ai(test_query)
print(f"2. AI Result:      {ai_intent} ({score*100:.2f}%) (Time: {time.time()-start:.4f}s)")
```

```python
from transformers import pipeline

# 1. Load a "Zero-Shot" Classifier (No API key needed)
# This model is specifically designed for logical extraction tasks.
ie_model = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text = "On July 10, Tim Cook announced that Apple acquired a London-based startup for $1 billion."

# 2. Define our "Structured" Labels
# We group these to handle NER, Relationships, and Events
ner_labels = ["Person", "Organization", "Location", "Date"]
event_labels = ["Business_Acquisition", "Public_Announcement", "Product_Launch"]

# 3. Extraction Logic
print("--- STRUCTURED EXTRACTION (OPEN SOURCE) ---")

# Part A: Event Extraction
event_result = ie_model(text, candidate_labels=event_labels)
print(f"EVENT: {event_result['labels'][0]} (Confidence: {round(event_result['scores'][0], 4)})")

# Part B: Relationship Extraction (Manual Prompting)
# We ask the model to verify a relationship
rel_labels = ["Tim Cook works for Apple", "Apple works for Tim Cook"]
rel_result = ie_model(text, candidate_labels=rel_labels)
print(f"RELATIONSHIP: {rel_result['labels'][0]}")

# Part C: Entity Check
# Checking if 'Apple' is an Organization
entity_check = ie_model("Apple", candidate_labels=ner_labels)
print(f"ENTITY TYPE (Apple): {entity_check['labels'][0]}")
```

```python
# First, run this in your terminal: pip install transformers torch
from transformers import pipeline

# 1. Load the Open-Source LLM (No API Key needed)
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

text_to_process = """
1. Industrial activities and fossil fuels drive climate change.
2. Deforestation reduces CO2 absorption.
3. Glaciers are melting and sea levels are rising.
4. Transitioning to renewable energy is essential.
"""

# 2. Categories defined in your prompt
categories = ["CAUSES", "EFFECTS", "SOLUTIONS"]

# 3. Execution Logic
print("--- STRUCTURED SUMMARY ---")

# We split the text into lines to categorize each one accurately
lines = [line.strip() for line in text_to_process.strip().split('\n') if line.strip()]
results = {"CAUSES": [], "EFFECTS": [], "SOLUTIONS": []}

for line in lines:
    # Remove the "1. " numbering for cleaner processing
    clean_text = line.split('. ', 1)[-1] if '. ' in line else line

    # The LLM determines which category the sentence belongs to
    prediction = classifier(clean_text, candidate_labels=categories)
    category = prediction['labels'][0]
    results[category].append(clean_text)

# 4. Final Output in your exact requested format
for cat in categories:
    print(f"\n- [{cat}]:")
    if results[cat]:
        for item in results[cat]:
            print(f"  * {item}")
    else:
        print("  * No data found.")
```

```python
# Run this first: !pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util

# 1. Load a pre-trained "Embedding" model
# This model converts sentences into math vectors.
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2. Input your two texts
text1 = "Industrial activities are a primary driver of climate change due to fossil fuel burning."
text2 = "The main cause of global warming is human industry and the use of coal and oil."

# 3. Compute the "Embeddings" (The math version of the text)
embedding1 = model.encode(text1, convert_to_tensor=True)
embedding2 = model.encode(text2, convert_to_tensor=True)

# 4. Calculate Similarity Score (0.0 to 1.0)
cosine_score = util.cos_sim(embedding1, embedding2)

print("--- PLAGIARISM CHECK ---")
print(f"Text A: {text1}")
print(f"Text B: {text2}")
print(f"\nSimilarity Score: {cosine_score.item():.4f}")

# 5. Interpretation
if cosine_score > 0.8:
    print("RESULT: HIGH SIMILARITY (Possible Plagiarism)")
elif cosine_score > 0.5:
    print("RESULT: MODERATE SIMILARITY (Paraphrased)")
else:
    print("RESULT: LOW SIMILARITY (Unique Content)")
```
