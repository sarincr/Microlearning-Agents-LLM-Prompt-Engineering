Query-Focused Summarization : The summary is generated based on a specific question or topic given in the prompt.

Focuses only on relevant information

Ignores unrelated details

```python
text = """
1. Industrial activities are a primary driver of climate change.
2. The burning of fossil fuels releases carbon dioxide into the air.
3. Deforestation reduces the planet's ability to absorb CO2.
4. Methane emissions from agriculture also contribute to warming.
5. Glaciers are melting at an alarming rate in the Arctic.
6. Sea levels are rising, threatening coastal cities worldwide.
7. Extreme weather patterns are becoming more frequent.
8. The loss of biodiversity is a major concern for scientists.
9. Oceans are becoming more acidic as they absorb carbon.
10. Global temperatures have risen significantly over the last century.
11. Increased greenhouse gases trap heat within the atmosphere.
12. Changes in solar radiation have a minor impact on climate.
13. Volcanic eruptions can cause temporary cooling effects.
14. Adaptation strategies are being developed by many nations.
15. Transitioning to renewable energy is essential for the future.
"""

response = model.generate_content(
    contents=f'''Query: What are the physical changes happening to the Earth? Summarize ONLY these from the text in 3 abstractive bullet points starting with '•'.
     Ignore causes like fuels or industry. Text: {text}'''
)

print("--- QUERY-FOCUSED SUMMARY (EARTH CHANGES) ---")
print(response.text)
```

Progressive / Iterative Summarization : Text is summarized step-by-step in multiple stages (chunk → summary → final summary).

Useful for long documents

Reduces token limits

Improves coherence

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Load the Open-Source Summarizer (BART)
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Helper function for summarization
def bart_summarize(text_input, max_len, min_len):
    inputs = tokenizer(text_input, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=max_len,
        min_length=min_len,
        early_stopping=True,
        do_sample=False # Explicitly set for reproducibility, matching original intent
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# 2. Your long text (Split into two chunks to demonstrate iteration)
chunk_1 = """
Industrial activities are a primary driver of climate change.
The burning of fossil fuels releases carbon dioxide into the air.
Deforestation reduces the planet's ability to absorb CO2.
Methane emissions from agriculture also contribute to warming.
"""

chunk_2 = """
Glaciers are melting at an alarming rate in the Arctic.
Sea levels are rising, threatening coastal cities worldwide.
Extreme weather patterns are becoming more frequent.
Oceans are becoming more acidic as they absorb carbon.
"""

# --- STAGE 1: Summarize Chunks ---
# We compress each chunk into a single sentence.
s1 = bart_summarize(chunk_1, max_len=30, min_len=10)
s2 = bart_summarize(chunk_2, max_len=30, min_len=10)

print(f"Intermediate 1: {s1}")
print(f"Intermediate 2: {s2}")

# --- STAGE 2: Final Synthesis ---
# We combine the two summaries and summarize them one last time.
final_input = s1 + " " + s2
final_summary = bart_summarize(final_input, max_len=40, min_len=15)

print("\n--- FINAL PROGRESSIVE SUMMARY ---")
print(f"• {final_summary.strip()}")
```

Structured Summarization : The summary follows a specific format or structure such as headings, tables, or categories.

Organized output

Useful for research or analysis

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load BART model
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = """
Industrial activities are a primary driver of climate change.
The burning of fossil fuels releases carbon dioxide into the air.
Deforestation reduces the planet's ability to absorb CO2.
Methane emissions from agriculture also contribute to warming.
Glaciers are melting at an alarming rate in the Arctic.
Sea levels are rising, threatening coastal cities worldwide.
Extreme weather patterns are becoming more frequent.
The loss of biodiversity is a major concern for scientists.
Oceans are becoming more acidic as they absorb carbon.
Global temperatures have risen significantly over the last century.
Increased greenhouse gases trap heat within the atmosphere.
Changes in solar radiation have a minor impact on climate.
Volcanic eruptions can cause temporary cooling effects.
Adaptation strategies are being developed by many nations.
Transitioning to renewable energy is essential for the future.
"""

sentences = text.strip().split("\n")

# Category storage
causes = []
impacts = []
solutions = []

# Keyword based categorization
for s in sentences:
    s_lower = s.lower()

    if any(word in s_lower for word in ["industrial", "fossil", "deforestation", "methane", "greenhouse", "solar", "volcanic"]):
        causes.append(s)

    elif any(word in s_lower for word in ["melting", "rising", "weather", "biodiversity", "acidic", "temperatures"]):
        impacts.append(s)

    elif any(word in s_lower for word in ["adaptation", "renewable"]):
        solutions.append(s)

# Convert list to text
causes_text = " ".join(causes)
impacts_text = " ".join(impacts)
solutions_text = " ".join(solutions)

print("\n--- STRUCTURED RESEARCH SUMMARY ---\n")

for category, content in {
    "CAUSES": causes_text,
    "IMPACTS": impacts_text,
    "SOLUTIONS": solutions_text
}.items():

    inputs = tokenizer(content, return_tensors="pt", max_length=1024, truncation=True)

    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=40,
        min_length=15,
        early_stopping=True
    )

    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    print("###", category)
    print(summary)
    print("-" * 30)
```
