```python
import nltk
from nltk import word_tokenize
from nltk.util import ngrams

# Sample text
text = "I love machine learning and natural language processing."

# Tokenize
tokens = word_tokenize(text.lower())

# Generate bigrams (2-word phrases)
bigrams = list(ngrams(tokens, 2))

# Define phrases to search
target_phrases = [("machine", "learning"), ("natural", "language")]

# Match using set intersection (no explicit loop)
matches = set(bigrams) & set(target_phrases)

print("Matched phrases:", matches)
```

```python
import spacy
from spacy.matcher import PhraseMatcher

# Load small English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "I love machine learning and natural language processing."

# Create PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

# Define phrases to search
phrases = ["machine learning", "natural language"]
patterns = [nlp.make_doc(phrase) for phrase in phrases]
print("patterns", patterns)
# Add patterns to matcher
matcher.add("TECH_PHRASES", patterns)
print("patterns", patterns)
# Process text
doc = nlp(text)

# Apply matcher
matches = matcher(doc)
print("matches", matches)
# Print results
for match_id, start, end in matches:
    span = doc[start:end]
    print("Matched phrase:", span.text)
```

```python
import spacy
from spacy.symbols import nsubj, VERB

nlp = spacy.load("en_core_web_sm")
doc = nlp("Autonomous cars shift insurance liability toward manufacturers")

# Finding a verb with a subject from below
verbs = set()
for possible_subject in doc:
    if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
        verbs.add(possible_subject.head)
print(verbs)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

# Inflectional Morphology (plural → singular, tense → base form)
print("Inflectional:")
print("cars →", lemmatizer.lemmatize("cars"))      # plural → car
print("running →", lemmatizer.lemmatize("running", pos="v"))  # verb tense → run

# Derivational Morphology (new word with different meaning)
from nltk.corpus import wordnet

print("\nDerivational:")
word = "happy"
synsets = wordnet.synsets(word)

for lemma in synsets[0].lemmas():
    if lemma.derivationally_related_forms():
        print(f"{lemma.name()} → {[d.name() for d in lemma.derivationally_related_forms()]}")
```

```python
import re
from collections import Counter

# Step 0.1: Load a large text corpus to build our language model (P(w))
# For simplicity, let's use a small excerpt. In a real application, you'd use
# a much larger corpus like Wikipedia, Project Gutenberg books, etc.

TEXT = """
This is a sample text for demonstrating spelling correction.
It contains common English words.
The quick brown fox jumps over the lazy dog.
Spelling correction is an interesting problem.
Let's see how it works.
"""

# Step 0.2: Preprocess the text to get a list of words
def words(text):
    return re.findall(r'[a-z]+', text.lower())

# Step 0.3: Build our "dictionary" and count word frequencies
# This will be used for P(w) - the prior probability
WORD_COUNTS = Counter(words(TEXT))
TOTAL_WORDS = sum(WORD_COUNTS.values())

print(f"Total unique words in corpus: {len(WORD_COUNTS)}")
print(f"Total words in corpus: {TOTAL_WORDS}")
print(f"Frequency of 'the': {WORD_COUNTS['the']}")
print(f"Frequency of 'spelling': {WORD_COUNTS['spelling']}")
```

```python
def P(word):
    "Probability of `word` occurring in our corpus."
    return WORD_COUNTS[word] / TOTAL_WORDS

print(f"P('the'): {P('the'):.6f}")
print(f"P('correction'): {P('correction'):.6f}")
print(f"P('xyz'): {P('xyz'):.6f}") # Will be 0 if not in corpus
```

```python
def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes    = [L + R[1:] for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:] for L, R in splits if R for c in letters]
    inserts    = [L + c + R for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word):
    "All edits that are two edits away from `word`."
    return {e2 for e1 in edits1(word) for e2 in edits1(e1)}

# Example:
print(f"Edits 1 for 'apple': {edits1('apple')}")
print(f"Number of edits 1 for 'apple': {len(edits1('apple'))}")
```

```python
def known(words_set):
    "The subset of `words_set` that appear in our `WORD_COUNTS`."
    return {w for w in words_set if w in WORD_COUNTS}

def candidates(word):
    "Generate probable correct spellings for `word`."
    # 1. Is the word itself in our dictionary?
    known_word = known({word})
    if known_word:
        return known_word

    # 2. Are there any words one edit away that are in our dictionary?
    known_edits1 = known(edits1(word))
    if known_edits1:
        return known_edits1

    # 3. Are there any words two edits away that are in our dictionary?
    known_edits2 = known(edits2(word))
    if known_edits2:
        return known_edits2

    # 4. If nothing found, return the original word (or an empty set, depending on desired behavior)
    return {word} # Or set() if you want to indicate no correction found
```

```python
def correction(word):
    "Most probable spelling correction for `word`."
    # Get all plausible candidates
    cands = candidates(word)

    # Find the candidate with the highest prior probability P(w)
    # max(iterable, key=function) returns the item with the highest value
    # based on the key function.
    best_candidate = max(cands, key=P)

    return best_candidate

# Let's test it!
print("\n--- Spelling Correction Examples ---")
print(f"'speling' -> {correction('speling')}")
print(f"'corrction' -> {correction('corrction')}")
print(f"'quik' -> {correction('quik')}")
print(f"'brwn' -> {correction('brwn')}")
print(f"'lazzy' -> {correction('lazzy')}")
print(f"'demnstrating' -> {correction('demnstrating')}")
print(f"'teh' -> {correction('teh')}") # Common typo
print(f"'apple' -> {correction('apple')}") # Already correct
print(f"'xyz' -> {correction('xyz')}") # Not in corpus, returns itself
print(f"'jumpss' -> {correction('jumpss')}") # Example of 2 edits if 'jumps' is in corpus
```
